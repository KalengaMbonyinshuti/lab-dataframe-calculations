## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here

aggregated_df = orders.groupby('CustomerID').agg({'amount_spent': 'sum'}).reset_index()
aggregated_df
```

```python
amount_spent_75th_percentile = aggregated_df['amount_spent'].quantile(0.75)
amount_spent_95th_percentile = aggregated_df['amount_spent'].quantile(0.95)

print(amount_spent_75th_percentile)
print(amount_spent_95th_percentile)
```

```python
aggregated_df['Customer_Group'] = 'Unclassified'
aggregated_df.loc[aggregated_df['amount_spent'] > amount_spent_95th_percentile, 'Customer_Group'] = 'VIP'
aggregated_df.loc[(aggregated_df['amount_spent'] >= amount_spent_75th_percentile) & (aggregated_df['amount_spent'] <= amount_spent_95th_percentile),'Customer_Group'] = 'Preferred VIP'
aggregated_df
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here

VIP_Customers = aggregated_df[aggregated_df['Customer_Group']=='VIP']
VIP_Customers['CustomerID']
VIP_Customers
```

```python
orders_vip_customer = orders[orders['CustomerID'].isin(VIP_Customers['CustomerID'])]
orders_vip_customer.head()
```

```python
orders_vip_customer = orders_vip_customer.groupby('Country').agg({'CustomerID': 'count'}).sort_values(by='CustomerID',                                                                                                  ascending=False)
orders_vip_customer
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
# Selecting Customers with VIP
VIP_PREF_Customers = aggregated_df[(aggregated_df['Customer_Group']=='VIP') | (aggregated_df['Customer_Group']=='Preferred VIP')]
# VIP_Customers['CustomerID']
VIP_PREF_Customers

```

```python
orders_vip_preferred_customer = orders[orders['CustomerID'].isin(VIP_PREF_Customers['CustomerID'])]
orders_vip_preferred_customer.head()
```

```python
orders_vip_pref_customer = orders_vip_preferred_customer.groupby('Country').agg({'CustomerID': 'count'}).sort_values(by='CustomerID',
                                                                                                     ascending=False)
orders_vip_pref_customer
```
